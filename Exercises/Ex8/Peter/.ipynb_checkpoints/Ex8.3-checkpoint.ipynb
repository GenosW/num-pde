```python
import netgen.gui
from netgen.csg import *
from ngsolve import *
from ngsolve.solvers import CG

import matplotlib.pyplot as plt
import numpy as np

from ngsolve.internal import visoptions
from ngsolve.internal import viewoptions
```

# AFW-preconditioner for Magnetostatic problem (Maxwells equation)

https://ngsolve.org/docu/nightly/i-tutorials/unit-2.4-Maxwell/Maxwell.html

\begin{equation}
\int{ν\ curl u \cdot curl v \ dx} + \int{ν \epsilon \ u \cdot v \ dx} = \int{f · v \ dx} \ \ \ , ∀v
\end{equation}

...

Solve with CG (rtol=$1e^{-8}$)

```python
def MakeGeometry():
    geometry = CSGeometry()
    box = OrthoBrick(Pnt(-1,-1,-1),Pnt(2,1,2)).bc("outer")

    core = OrthoBrick(Pnt(0,-0.05,0),Pnt(0.8,0.05,1))- \
           OrthoBrick(Pnt(0.1,-1,0.1),Pnt(0.7,1,0.9))- \
           OrthoBrick(Pnt(0.5,-1,0.4),Pnt(1,1,0.6)).maxh(0.2).mat("core")
    
    coil = (Cylinder(Pnt(0.05,0,0), Pnt(0.05,0,1), 0.3) - \
            Cylinder(Pnt(0.05,0,0), Pnt(0.05,0,1), 0.15)) * \
            OrthoBrick (Pnt(-1,-1,0.3),Pnt(1,1,0.7)).maxh(0.2).mat("coil")
    
    geometry.Add ((box-core-coil).mat("air"))
    geometry.Add (core)
    geometry.Add (coil)
    return geometry

ngmesh = MakeGeometry().GenerateMesh(maxh=0.5)
mesh = Mesh(ngmesh)

#ngsglobals.msg_level = 5

fes = HCurl(mesh, order=1, dirichlet="outer")

u = fes.TrialFunction()
v = fes.TestFunction()

mur = { "core" : 1000, "coil" : 1, "air" : 1 }
mu0 = 1.257e-6

nu_coef = [ 1/(mu0*mur[mat]) for mat in mesh.GetMaterials() ]
print ("nu_coef=", nu_coef)
```

```python
blocks = []
freedofs = fes.FreeDofs()
for v in mesh.vertices:
    vdofs = set()
    for el in mesh.vertices:
        vdofs |= set(d for d in fes.GetDofNrs(el) if freedofs[d])
    blocks.append(vdofs)
```

```python
num_jac = []
err_jac = []#np.zeros([10000,14])
num_block = []
err_block = []# np.zeros([10000,14])
temp = []
n = 1
rtol = 1e-8




epsilon = [10**(-j) for j in range(7)]
for eps in epsilon:
    print("epsilon: ",eps)
    
    nu = CoefficientFunction(nu_coef)
    u = fes.TrialFunction()
    v = fes.TestFunction()

    a = BilinearForm(fes, symmetric=True)
    a += nu*curl(u)*curl(v)*dx + eps*nu*u*v*dx

    # 1. the jacobi preconditioner
    c = Preconditioner(a, type="local")
    c.Update()
    # 2. the direct inverse 
    # c = Preconditioner(a, type="direct")
    
    f = LinearForm(fes)
    f += CoefficientFunction((y,0.05-x,0)) * v * dx("coil")
    
    blocks = []
    freedofs = fes.FreeDofs()
    for v in mesh.vertices:
        vdofs = set()
        for el in mesh.vertices:
            vdofs |= set(d for d in fes.GetDofNrs(el) if freedofs[d])
        blocks.append(vdofs)

    gfu_jac = GridFunction(fes,name = "u_jac")
    gfu_AFW = GridFunction(fes,name = "u_AFW")
    gfu = GridFunction(fes,name = "u")

    # stores the error and iteration number
    data_jac = []
    data_block = []
    # needed for this...
    callback_jac = lambda k_jac,r_jac: data_jac.append((k_jac,r_jac))
    callback_block = lambda k_block,r_block: data_block.append((k_block,r_block))

    with TaskManager():
        a.Assemble()    
        f.Assemble()

        solvers.CG(mat=a.mat, pre=c.mat, sol = gfu_jac.vec, rhs = f.vec, tol=rtol, maxsteps = 10000, callback = callback_jac, printrates = False)
        
        # Use your AFW preconditioner instead of the Jacobi
        # 3. AFW
        AFW = a.mat.CreateBlockSmoother(blocks)
        solvers.CG(mat=a.mat, pre=AFW, sol = gfu_AFW.vec, rhs = f.vec, tol =rtol, maxsteps = 10000, callback = callback_block, printrates = False)

    num_jac_loc = []
    err_jac_loc = []
    num_block_loc = []
    err_blockloc = []
    for dat in data_jac:        
        num_jac_loc.append(dat[0])
        err_jac_loc.append(dat[1])

    for dat in data_block:
        num_block_loc.append(dat[0])
        err_block_loc.append(dat[1])
        
    num_jac.append(num_jac_loc)
    err_jac.append(err_jac_loc)
    num_block.append(num_block_loc)
    err_block.append(err_blockloc)

visoptions.clipsolution = 'vec'
viewoptions.clipping.ny= -1
viewoptions.clipping.enable = 1
Draw(gfu_jac.Deriv(), mesh, "B-field_jac", draw_surf=False))
Draw(gfu_AFW.Deriv(), mesh, "B-field_AFW", draw_surf=False))
```

```python
plt.figure(figsize=(10,8))
for i, eps in enumerate(epsilon):
    print(eps)
    plt.loglog(num_jac[i],err_jac[i], label=str(eps))
    
plt.legend()
```

```python
plt.figure(figsize=(10,8))
for i, eps in enumerate(epsilon[:1]):
    print(eps)
    plt.loglog(num_block[i],err_block[i], label=str(eps))
    
plt.legend()
```
